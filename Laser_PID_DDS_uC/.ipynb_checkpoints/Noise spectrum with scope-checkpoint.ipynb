## Scope measurements

```python
# list available devices
rm = pyvisa.ResourceManager()
print(rm.list_resources())
```

```python
# open connections to devices
scope = rm.open_resource('USB0::0x1AB1::0x04CE::DS1ZA204619142::INSTR')
```

```python
def scope_measure(timebase):
    # set up the scope
    scope.write(f':TIM:SCAL {timebase}')
    scope.write(f':TIM:OFFS {6*timebase}')
    
    # trigger the scope
    scope.write(':SING')
    scope.write(':TFOR')

    # wait for measurement to complete
    while scope.query('TRIG:STAT?').strip() != "STOP":
        time.sleep(.25)
        
def get_waveform(max_pts=250000):
    # set the scope data format
    scope.write(':WAV:SOUR CHAN1')
    scope.write(':WAV:MODE RAW')
    scope.write(':WAV:FORM BYTE')
    
    # get & set waveform parameters
    mdepth = int(scope.query(':ACQ:MDEP?'))
    x_increment = float(scope.query(':WAV:XINC?'))
    x_origin    = float(scope.query(':WAV:XOR?'))
    x_reference = float(scope.query(':WAV:XREF?'))
    y_increment = float(scope.query(':WAV:YINC?'))
    y_origin    = float(scope.query(':WAV:YOR?'))
    y_reference = float(scope.query(':WAV:YREF?'))

    # get waveform data from scope
    curve = np.zeros(mdepth)
    for i in tqdm(range(int(np.ceil(mdepth/max_pts)))):
        # read a chunk from the scope
        i0, i1 = i*max_pts, i*max_pts + min(max_pts, mdepth - i*max_pts)
        scope.write(f':WAV:STAR {i0+1}')
        scope.write(f':WAV:STOP {i1}')
        scope.write(':WAV:DATA?')
        raw_data = scope.read_raw()[11:-1]
        
        # put the data into the numpy array
        curve[i0:i1] = np.frombuffer(raw_data, dtype=np.uint8).astype(float)

    # convert raw data to voltage and time
    volts = (curve - y_reference - y_origin) * y_increment
    t = np.linspace(x_origin, x_origin + x_increment*len(volts), len(volts))
    
    return t, volts

def avg_spectrum(t, volts, n_avg):
    spectrum_arr = []
    for volts_chunk in np.split(volts, n_avg):
        freq, spectrum_chunk = periodogram(volts_chunk, 1/(t[1]-t[0]))
        spectrum_arr.append( spectrum_chunk )
    spectrum = 10*np.log10(np.mean(spectrum_arr, axis=0))
    return freq, spectrum

def plot_spectrum(freq, spectrum, n_avg):
    # plot the spectrum
    plt.semilogx(freq, spectrum, color='black', lw=1)
    plt.xlabel("frequency [Hz]")
    plt.ylabel("power [dB]")
    plt.title(f'n_avg = {n_avg}')
    plt.grid()

def get_spectrum(timebase=0.05, n_avg=1):
    scope_measure(timebase)
    t, volts = get_waveform()
    plot_spectrum(*avg_spectrum(t, volts, n_avg), n_avg)
    return t, volts
```

## Plotting data from moku:lab

```python
def load_traces(f_str, folder):
    freq, moku, _ = np.genfromtxt(f"{folder}/moku{f_str}_Traces.csv", delimiter=',', autostrip=True, dtype=float, skip_header=7).T
    freq, preamp, _ = np.genfromtxt(f"{folder}/preamp{f_str}_Traces.csv", delimiter=',', autostrip=True, dtype=float, skip_header=7).T
    freq, locked, _ = np.genfromtxt(f"{folder}/locked{f_str}_Traces.csv", delimiter=',', autostrip=True, dtype=float, skip_header=7).T
    freq, unlocked, _ = np.genfromtxt(f"{folder}/unlocked{f_str}_Traces.csv", delimiter=',', autostrip=True, dtype=float, skip_header=7).T
    return freq, moku, preamp, locked, unlocked

def join_traces(folder="data/setpoint_3000"):
    # for storing result
    freq, moku, preamp, locked, unlocked = [], [], [], [], []

    for f0,f1,f_str in [[0,1e3,"1kHz"], [1e3,1e4,"10kHz"], [1e4,1e5,"100kHz"], [1e5,1e6,"1MHz"]]:
        # load data
        freq_c, moku_c, preamp_c, locked_c, unlocked_c = load_traces(f_str, folder)

        # find indices of frequency range start/stop points
        f0_idx = np.argmin(np.abs(freq_c-f0))
        f1_idx = np.argmin(np.abs(freq_c-f1))

        # stitch together the chunk with the rest of the data
        freq.extend(freq_c[f0_idx:f1_idx])
        moku.extend(moku_c[f0_idx:f1_idx])
        preamp.extend(preamp_c[f0_idx:f1_idx])
        locked.extend(locked_c[f0_idx:f1_idx])
        unlocked.extend(unlocked_c[f0_idx:f1_idx])
        
    return freq, moku, preamp, locked, unlocked
```

```python
freq, moku, preamp, locked, unlocked = join_traces(folder="data/setpoint_37000K0.1")
fig, ax = plt.subplots(figsize=(9,7))
ax.semilogx(freq, moku, color='black', lw=2, label="moku:lab")
ax.semilogx(freq, preamp, color='gray', lw=1, label="preamp")
ax.semilogx(freq, unlocked, color='blue', lw=2, label="unlocked")
ax.semilogx(freq, locked, color='red', lw=2, label="locked")
ax.set_xlabel("frequency [Hz]")
ax.set_ylabel("power [dBm/Hz]")
ax.set_title(f"setpoint = 3000, n_avg = 100")
ax.legend()
# ax.set_xlim([5,1e3])
ax.grid()
```
